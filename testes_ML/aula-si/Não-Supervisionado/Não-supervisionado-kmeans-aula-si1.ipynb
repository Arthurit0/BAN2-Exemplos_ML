```python
import pandas as pd
from sklearn.cluster import KMeans
```

###Segmentação de Clientes: Uma empresa pode utilizar o K-Means para segmentar seus clientes em grupos com base em suas características demográficas, comportamentais ou de compra. Esses grupos podem ajudar a empresa a personalizar estratégias de marketing direcionadas a cada segmento.

```python
# Criar um DataFrame com dados de clientes
data = {'idade': [25, 45, 30, 50, 60, 20],
        'gasto': [100, 300, 200, 150, 50, 250]}
df = pd.DataFrame(data)
df
```

```python
qt_grupos = 2

# Aplicar o algoritmo de K-Means para agrupar os clientes
kmeans = KMeans(n_clusters=qt_grupos)
```

K-Means clustering.

class sklearn.cluster.KMeans(n_clusters=8, *, init='k-means++', n_init='warn', max_iter=300, tol=0.0001, verbose=0, random_state=None, copy_x=True, algorithm='lloyd')[source]

```python
kmeans.fit(df)

# Obter os rótulos dos clusters
labels = kmeans.labels_
```

```python
df['grupo'] = labels
df
```
